```python
import cv2
import os





data_path = "Generative-Adversarial-Networks--GANs-main/Data"


output_path = "resized_grayscale_images"  

new_size = (512, 512)


# Get all image paths in the data folder
image_paths = [os.path.join(data_path, f) for f in os.listdir(data_path) if os.path.isfile(os.path.join(data_path, f))]

# Loop through each image path
for image_path in image_paths:
    try:
        # Read the image
        image = cv2.imread(image_path)

        # Check if image is read successfully
        if image is not None:
            # Used LLM to find the function for greyscale
            grayscale_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

            # Resize the image
            resized_image = cv2.resize(grayscale_image, new_size)

            # Create the full output path with unique filename
            output_filename = os.path.join(output_path, f"resized_grayscale_{os.path.basename(image_path)}")

            # Save the resized image
            cv2.imwrite(output_filename, resized_image)

            print(f"Image resized, converted to grayscale, and saved: {output_filename}")
        else:
            print(f"Error reading image: {image_path}")
    except Exception as e:
        print(f"Error processing image: {image_path} - {e}")

print("All images processed!")
```
